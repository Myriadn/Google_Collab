# **Import Library dan Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math

dataset = pd.read_csv("Daftar_Gaji.csv", sep = ";")
dataset.head()
```

# **Jumlah Dataset**

```python
print("Jumlah Datasetnya : " + str(len(dataset.index)))
```

# **Analisis Deskriptif otomatis pada dataset**

```python
dataset.describe()
```

# **Menampilkan Grafik Plot**

```python
dataset.plot(x="Tahun_bekerja", y="Gaji", style="o")
plt.title("Grafik Gaji Berdasarkan Tahun Bekerja")
plt.xlabel("Tahun Bekerja")
plt.ylabel("Gaji")
plt.show
```

# **Import Linear Regresi dan buat objek Regressor**

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

# **Prediksi Gaji Karyawan secara Manual**

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
# Perhitungan manual dengan rumus :
# y = intercept + coef * tahun_kerja
26780.09915062818 + 9312.57512673 * 1.5
```

# **Prediksi Gaji Karyawan secara Otomatis**

```python
y_pred = regressor.predict(x_test)
DataFrame = pd.DataFrame({'Data Asli': y_test, 'Data Di Prediksi': y_pred})
DataFrame
```

```python
# Visualisasi hasil Test Set
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Gaji vs Tahun Bekerja (Test Set)')
plt.xlabel('Lama Bekerja (Tahun)')
plt.ylabel('Gaji')
plt.show()
```

```python
# Visualisasi Hasil Training set
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Gaji vs Tahun Bekerja (Training Set)')
plt.xlabel('Lama Bekerja (Tahun)')
plt.ylabel('Gaji')
plt.show()
```
